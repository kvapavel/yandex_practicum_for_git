Зафиксируем псевдослучайность для всех используемых в проекте алгоритмов:

```python
rand_st = 12345
```

Подготовим фичи и целевые признаки обучающей, валидационной и тестовой выборок.

```python
target = data['exited']
features = data.drop('exited', axis=1)

features = pd.get_dummies(features, drop_first=True)

target_train, target_valid, features_train, features_valid =\
train_test_split(target, features, test_size=0.4, random_state=rand_st, stratify=target)

target_valid, target_test, features_valid, features_test =\
train_test_split(target_valid, features_valid, test_size=0.5, random_state=rand_st, stratify=target_valid)
```

В итоге получили следующее соотношение:

|train |valid | test |
|------|------|------|
|60%   | 20%  | 20%  |

Масштабирование данных.


Прежде чем приступать к моделированию, нужно выполнить «центровку» и «стандартизацию» данных путём их масштабирования. Масштабирование выполняется из-за того, что разные величины выражены в разных единицах измерения. Эта процедура позволяет организовать «честную схватку» между признаками при определении их важности.

```python
numeric = ['creditscore', 'age', 'tenure', 'balance', 'numofproducts', 'estimatedsalary']

scaler = StandardScaler()
scaler.fit(features_train[numeric])
features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])



print(features_train.shape)
print(features_valid.shape)
print(features_test.shape)
```

Для решения нашей задачи далее рассмотрим следующие изученные модели:

* дерево решений / decision tree
* случайный лес  / random forest
* логистическую регрессию / logistic regression

### Логистическая регрессия

```python
model_lgr = LogisticRegression(solver='liblinear', random_state=rand_st)
model_lgr.fit(features_train, target_train)
predicted_valid = model_lgr.predict(features_valid)

print(f'F1-мера на валидационной выборке: {f1_score(target_valid, predicted_valid)}')
```

```python
probabilities_valid = model_lgr.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]


print('ROC_AUC:', roc_auc_score(target_valid, probabilities_one_valid))
```

### Дерево решений

```python
scores_data = pd.DataFrame() # создаём пустой датафрейм
```

```python
%%time

best_model_dt = None
best_result = 0
best_max_depth = 0

for depth in range(1, 51):
    model = DecisionTreeClassifier(random_state=rand_st, max_depth=depth)
    model.fit(features_train, target_train)
    
    predicted_train = model.predict(features_train)
    predicted_valid = model.predict(features_valid)
    train_score = f1_score(predicted_train, target_train)
    valid_score = f1_score(predicted_valid, target_valid)
    
    temp_score_data = pd.DataFrame({'max_depth': [depth], 'train_score': [train_score], 'valid_score': [valid_score]})
    scores_data = pd.concat((scores_data,temp_score_data),axis=0)
    
    if valid_score > best_result:
        best_model_dt = model
        best_result = valid_score
        best_max_depth = depth
        
        
print("F1 лучшей модели:", best_result)
print("Глубина дерева лучшей модели:", best_max_depth)
print()
```

```python
probabilities_valid = best_model_dt.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]


print('ROC_AUC:', roc_auc_score(target_valid, probabilities_one_valid))
```

```python
scores_data_long = pd.melt(scores_data, id_vars = ['max_depth'], value_vars = ['train_score','valid_score'],\
                           var_name = 'set_type', value_name = 'score')
```

Посмотрим как меняется точность на train и validation датасете, при изменении гиперпараметров.

```python
sns.lineplot(x='max_depth', y='score', hue='set_type', data=scores_data_long);
```

Наглядно видим, что после max_depth=7 точность на train выборке растёт, а на valid - выборке точность падает. Из этого делаем вывод, что модель начинает переобучаться.

### Увеличение выборки

Поскольку у нас имеется дисбаланс классов. И в меньшинстве у нас класс 1, то для борьбы с данным дисбалансом, увеличим данный класс.

```python
#функция для увеличения класса 1.
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=rand_st)
    
    return features_upsampled, target_upsampled
```

### Уменьшение выборки

Теперь для балансировки классов наоборот уменьшим выборку, за счёт класса 0.

```python
#функция для уменьшения класса 0.
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=rand_st)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=rand_st)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=rand_st)
    
    return features_downsampled, target_downsampled
```
